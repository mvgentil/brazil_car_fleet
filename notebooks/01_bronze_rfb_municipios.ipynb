```python
%sql
USE CATALOG brazil_car_fleet;
USE SCHEMA bronze;
```

```python
%sql
DROP TABLE IF EXISTS rfb_municipios;
```

```python
%sql

CREATE TABLE rfb_municipios AS
SELECT * FROM read_files(
  '/Volumes/brazil_car_fleet/raw_data/rfb_municipios_raw/municipios.csv',
  header => true,
  format => 'csv',
  delimiter => ';',
  schema => 'id_municipio int, cd_municipio string, nm_municipio_upper string, nm_municipio_formatted string, sigla_uf string',
  encoding => 'ISO-8859-1'
);
```
